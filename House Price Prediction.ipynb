## Splitting training and testing data

```python
from sklearn.model_selection import train_test_split
```

```python
train_x,test_x,train_y,test_y=train_test_split(housing.iloc[:,:-1],housing.iloc[:,[-1]],test_size=0.3)
```

## Selecting a model 

```python
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import cross_val_score 
```

```python
def give_model_report(model,name):
    print("========= {} ===========".format(name))
    scores = cross_val_score(model, housing.iloc[:,:-1], housing.iloc[:,[-1]], cv=5)
    print("Scores:", scores)
    print("Mean: ", scores.mean())
    print("Standard deviation: ", scores.std())
```

```python
give_model_report(LinearRegression(),"Linear Regression")
```

```python
give_model_report(KNeighborsRegressor(),"KNeighbors Regression")
```

```python
give_model_report(DecisionTreeRegressor(),"Decision Tree Regression")
```

```python
give_model_report(RandomForestRegressor(),"Random Forest Regression")
```

## improving selected model

```python
from sklearn.ensemble import RandomForestRegressor
model=RandomForestRegressor()
```

```python
model.fit(train_x,train_y)
```

```python
model.score(test_x,test_y)
```

```python
mymodel2 = RandomForestRegressor(n_estimators=10)
mymodel2.fit(train_x,train_y)
mymodel2.score(test_x,test_y)
```

```python
mymodel3 = RandomForestRegressor(n_estimators=25) 
mymodel3.fit(train_x,train_y)
mymodel3.score(test_x,test_y)
```
